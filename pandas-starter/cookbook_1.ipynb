```python
complaints_df["Complaint Type"][:10].value_counts().plot(kind='bar', figsize=(15, 10)) # bar
```

```python
# Plotting and filtering row
complaints_df.loc[(complaints_df["Complaint Type"]== "Rodent") | (complaints_df["Complaint Type"]== "Illegal Parking"), "Complaint Type"].value_counts().plot(kind="bar") 
```

```python
# get the percentage of Illegal Parking complaints from all complaints
parking_complaints = complaints_df[complaints_df["Complaint Type"] == "Illegal Parking"]["Borough"].value_counts()
all_complaints = complaints_df["Borough"].value_counts()
(parking_complaints / all_complaints).plot(kind="bar")
```

```python
# Grouping and aggregates
bikes_df = pd.read_csv("bikes.csv", sep=";", encoding="latin-1", parse_dates=["Date"], dayfirst=True, index_col="Date")
bikes_df
```

```python
bikes_df['Berri 1'].plot()
```

```python
# Dataframe adding value automatically adds value for all rows in dataframe
bikes_df["newcol"] = "row values"
bikes_df
```

```python
# With filtered row and new column
bikes_df.loc[bikes_df["Berri 1"] == 35,"new column"] = "Cal 35"
bikes_df
```

```python
# Copy as Dataframe not (Series) bikes_df["Berri 1"].copy()
berri_bikes = bikes_df[["Berri 1"]].copy()

# Using loc to target location for row value and adding new column
berri_bikes.loc[:, "Weekday"] = berri_bikes.index.weekday

# Without loc
berri_bikes["Weekday"] = berri_bikes.index.weekday
berri_bikes
```

```python
# Get all sum by weekday
berri_bikes.groupby("Weekday").aggregate(sum)
```

```python
# Change index
berri_bikes_weekday_data = berri_bikes.groupby("Weekday").aggregate(sum)
berri_bikes_weekday_data.index = ["Mon", "Tue", "Wed", "Thurs", "Fri", "Sat", "Sun"]
berri_bikes_weekday_data
```

```python
berri_bikes_weekday_data.plot()
```

```python
berri_bikes_weekday_data.plot(kind='bar')
```
