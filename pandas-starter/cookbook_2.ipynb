```python
# Weather Data
import pandas as pd

weather_df = pd.read_csv("weather.csv", parse_dates=["Date/Time"], index_col="Date/Time")
weather_df["Temp (C)"].plot(figsize=(10,6))
```

```python
# Read from a url that returns a CSV file format
url="http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year=2012&Month=3&timeframe=1&submit=Download+Data"
weather_mar2012 = pd.read_csv(url, index_col='Date/Time (LST)', parse_dates=True)
weather_mar2012
```

```python
weather_mar2012["Temp (°C)"].plot(figsize=(15, 10))
```

```python
# Other way of renaming columns
# Must be exact list length with actual column
weather_mar2012.columns = [
    "Longitude (x)",
    "Latitude (y)",        
    "Station Name ",
    "Climate ID", 
    u'Year',
    u'Month',
    u'Day',
    u'Time',
    u'Temp (C)', 
    u'Temp Flag',
    u'Dew Point Temp (C)',
    u'Dew Point Temp Flag', 
    u'Rel Hum (%)',
    u'Rel Hum Flag',
     "Precip. Amount (mm)",
     "Precip. Amount Flag",
    u'Wind Dir (10s deg)',
    u'Wind Dir Flag', 
    u'Wind Spd (km/h)',
    u'Wind Spd Flag',
    u'Visibility (km)',
    u'Visibility Flag',
    u'Stn Press (kPa)',
    u'Stn Press Flag',
    u'Hmdx',
    u'Hmdx Flag',
    u'Wind Chill', 
    u'Wind Chill Flag',
    u'Weather']
weather_mar2012.head()
```

```python
# Dropping columns that have null values
# axis=1 - Column, default is row
# how="any" - Has any null values 
weather_mar2012_drop = weather_mar2012.dropna(axis=1, how="any")
weather_mar2012_drop.head()
```

```python
# Drop additional columns
weather_df_new = weather_mar2012_drop.drop(["Year", "Month", "Day", "Time"], axis=1)
weather_df_new
```

```python
# Using mean to plot temp
temp_df = weather_df_new[["Temp (C)"]].copy()
temp_df.loc[:, "Hour"] = weather_mar2012.index.hour
temp_df.groupby("Hour").aggregate("mean").plot()
```
