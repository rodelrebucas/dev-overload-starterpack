```python
# Getting subsets from dataframe
fname_profession = users[["firstname", "profession"]]
fname_profession
```

```python
# filtering
doctor = users[users["profession"]== "doctor"]
doctor
```

```python
doctor_developer  = users[users["profession"].isin(["doctor", "developer"])]
doctor_developer
```

```python
doctor_developer  = users[(users["profession"] == "doctor") | (users["profession"] == "developer")]
doctor_developer
```

```python
# Selecting rows and filtering
f_lname = users.loc[(users["profession"]=="firefighter") | (users["profession"] == "doctor"),["firstname", "lastname", "profession"]]
f_lname
```

```python
# selecting specific no. of rows and cols
# Row 2 to 5 and Col 2 to 4
users.iloc[2:5,2:4]
```

```python
# Replacing values with either loc or iloc
# Replace lastname
users.iloc[2:5,2] = "RRR"
users["lastname"]
```
